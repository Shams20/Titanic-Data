# EDA 

```python
df['Age'].fillna(df['Age'].mode()[0],inplace=True)
```

```python
df.Age.isnull().sum()
```

```python
df['Sex']=pd.get_dummies(df['Sex'],drop_first=True)
```

```python
df['Embarked']=pd.get_dummies(df['Embarked'],drop_first=True)
```

# Creating Model

```python
x=df.drop(['Name','Ticket','Cabin','PassengerId'],axis=1)
y=df['Survived']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=0)
```

```python
print('x_train',len(x_train))
print('y_train',len(y_train))
print('x_test',len(x_test))
print('y_test',len(y_test))
```

# Prediction

```python
from sklearn.neighbors import KNeighborsClassifier

clf = KNeighborsClassifier(n_neighbors=3)
```

```python
df.dtypes
```

```python
x.head(5)
```

```python
clf=clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix
```

```python
print('accuracy_score of x_test Data:',accuracy_score(y_test,y_pred)*100)
print('\n','classification_report of x_test Data:','\n','\n',classification_report(y_test,y_pred)*100)
```

```python
print(confusion_matrix(y_test,y_pred))
```
